```python
from dateutil.relativedelta import *
import json 
import boto3 as b 
import base64
from faker import Faker
import time 
fake = Faker()
stream_name = 'kinesis-stream-json-files'
kinesis_client = b.client('kinesis')
```

```python
input_jsons = []
for batch_round in range(1,4):
    print(f'Batch Count {batch_round}\n')
    upload_count = 0
    for _ in range(0,fake.random_int(1,10)): 
        input_json = {}
        input_json['partition_key'] = (fake.uuid4())
        input_json['customer_id'] = (fake.random_int(min=1000,max=9999))
        input_json['name_prefix'] = (fake.prefix())
        input_json['first_name'] = (fake.first_name())
        input_json['last_name'] = (fake.last_name())
        input_json['salary'] = ('$'+fake.numerify('$$000'))
        input_json['occupation'] = (fake.random_choices(['Laywer','Doctor','Engineer','Programmer','Artist','Singer','Dancer','Driver','Pilot','Gamer'],length=1))
        input_json['date_of_birth'] = (fake.date_of_birth(minimum_age=18, maximum_age=65)).strftime('%d-%m-%Y')
        input_json['gender'] = (fake.random_choices(['Male','Female'],length=1))
        input_json['spending_score'] = (fake.random_int(min=0,max=100))
        input_json['customer_since_years'] = (fake.random_int(min=1,max=10))
        input_json['family_size'] = (fake.random_int(min=1,max=7))
        json_dataset = json.dumps(input_json)
        response = kinesis_client.put_record(StreamName=stream_name, Data=json_dataset, PartitionKey = input_json['partition_key'] )
        upload_count = upload_count + 1
        print(f"Uploaded Customer ID `{input_json['customer_id']}` to Shard ID `{response['ShardId']}` with Sequence Number `{response['SequenceNumber']}`")
    print(f'\nUploaded {upload_count} customer details to the stream')
    print('Sleep - 5 Seconds\n')
    time.sleep(5)
```
